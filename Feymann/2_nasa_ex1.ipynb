```python
import numpy as np
import re
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy import stats
from matplotlib.patches import Polygon
import math
from copy import copy
```

```python
def d(x):
    first = np.sqrt(np.power(x,2) + np.power(28.5,2))
    second = np.sqrt(np.power((x-19),2) + np.power(17,2))
    return np.round(first + second, 5)

d_f = np.vectorize(d)
```

```python
d(0),d(3), d(9.5), d(11.5),d(11.9), d(19)
```

```python
figsize=(20, 5)
fontsize_x=20.0
x_ticks_size=19
fontsize_y=20
y_ticks_size=19
linewidth=1

x = np.linspace(0,19,1000)
y1 = d_f(x)
fig, ax = plt.subplots(figsize=figsize)
plt.plot(x,y1, color="k");
i = np.argmin(y1)
print("x = {:.5f} | y = {:.5f}".format(x[i], y1[i]))
```

```python
def time(x):
    first = np.sqrt(np.power(x,2) + np.power(28.5,2)) / 8.9
    second = np.sqrt(np.power((x-19),2) + np.power(17,2)) / 5.5
    return np.round(first + second, 5)

time_f = np.vectorize(time)
```

```python
time(0),time(3), time(9.5), time(11.5),time(11.9), time(19)
```

```python
figsize=(20, 5)
fontsize_x=20.0
x_ticks_size=19
fontsize_y=20
y_ticks_size=19
linewidth=1

x = np.linspace(0,19,1000)
y1 = time_f(x)
fig, ax = plt.subplots(figsize=figsize)
plt.plot(x,y1, color="k");
i = np.argmin(y1)
print("x = {:.5f} | y = {:.5f}".format(x[i], y1[i]))
```
