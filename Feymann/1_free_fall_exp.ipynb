```python
figsize=(10, 5)
fontsize_x=20.0
x_ticks_size=19
fontsize_y=20
y_ticks_size=19
linewidth=1


def x(t):
    return 20*t + 0.5*-9.8*(t**2)

x_f = np.vectorize(x)
t = np.linspace(0.,6.0,10)
y = x_f(t)
fig, ax = plt.subplots(figsize=figsize)
plt.plot(t,y, color="k");
```

```python
pi = np.pi
x = np.linspace(0, 5*(2*pi), 500)
figsize=(20, 5)
fontsize_x=20.0
x_ticks_size=19
fontsize_y=20
y_ticks_size=19
linewidth=1

y = np.sin(x)
fig, ax = plt.subplots(figsize=figsize)
plt.plot(x,y, color="k");
plt.axvline(x=0.5*pi, color="r");
plt.axvline(x=pi, color="b");
plt.axvline(x=(3/2)*pi, color="g");
plt.axvline(x=2*pi, color="y");
```
