```python
x = np.linspace(0,math.radians(90),100)
xx = [math.degrees(i) for i in x]
y = np.sin(x)
figsize=(5, 5)
fig, ax = plt.subplots(figsize=figsize)
plt.plot(xx,y, color="k");

z = np.arcsin(y)
w = [math.degrees(i) for i in z]
fig, ax = plt.subplots(figsize=figsize)
plt.plot(y,w, color="r");
print("sin and arcsin\n")
```

```python
def yf(t):
    return 0.683 * np.power(t,2)


def test(b):
    return b/np.sqrt(1 + np.power(b,2))


def hf(t):
    b_sqr = np.power(yf(t),2)
    a_sqr = np.power(1000,2)
    return np.sqrt(a_sqr + b_sqr)

def F(t):
    ratio = yf(t)/hf(t)
    return np.arcsin(ratio)
```

```python
x = np.linspace(0,90,1000)
y1 = F(x)
y = [math.degrees(i) for i in y1]
figsize=(20, 5)
fig, ax = plt.subplots(figsize=figsize)
plt.plot(x,y, color="k");
print("time x angle plot")
```

```python
def delF_delt(t, e=0.0001):
    return (F(t+e) - F(t))/e

x = np.linspace(0.0001,0.001,3)
for i in x[::-1]:
    print(delF_delt(10,e=i))
```

```python
x = np.linspace(0,90,1000)
y = delF_delt(x)
figsize=(20, 5)
fig, ax = plt.subplots(figsize=figsize)
plt.plot(x,y, color="k");
i = np.argmax(y)
theta = math.degrees(F(x[i]))
print("del angle / del time plot\n")
```

```python
print("max: t = {:.5f} | theta = {:.5f}º".format(x[i], theta))
print("max: t = {:.5f} | theta = {:.5f} radians".format(x[i], F(x[i])))
```
