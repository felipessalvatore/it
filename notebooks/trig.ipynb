```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
figsize=(10, 5)
fontsize_x=20.0
x_ticks_size=19
fontsize_y=20
y_ticks_size=19
linewidth=1

x = np.linspace (0.01,0.99,10000)
y = x/(1-x)
fig, ax = plt.subplots(figsize=figsize)
plt.plot(x,y, color="k");
plt.xlabel('p');
plt.ylabel('ods');
```

```python
a1 = np.random.choice(y,5000)
n, bins, patches = plt.hist(a1, 50, density=True, facecolor='g', alpha=0.75)
plt.xlabel('Ods')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
figsize=(10, 5)
fontsize_x=20.0
x_ticks_size=19
fontsize_y=20
y_ticks_size=19
linewidth=1

x = np.linspace (0.01,0.99, 100)
y = np.log(x/(1-x))
fig, ax = plt.subplots(figsize=figsize)
plt.plot(x,y, color="k");
plt.xlabel('p');
plt.ylabel('log(ods)');
```

```python
pi = np.pi
x = np.linspace(0, 5*(2*pi), 500)
figsize=(20, 5)
fontsize_x=20.0
x_ticks_size=19
fontsize_y=20
y_ticks_size=19
linewidth=1

y = np.sin(x)
fig, ax = plt.subplots(figsize=figsize)
plt.plot(x,y, color="k");
plt.axvline(x=0.5*pi, color="r");
plt.axvline(x=pi, color="b");
plt.axvline(x=(3/2)*pi, color="g");
plt.axvline(x=2*pi, color="y");
```
